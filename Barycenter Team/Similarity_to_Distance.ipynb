```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Read the adj matrix into a dataframe
df = pd.read_csv(r"C:\Users\praje\OneDrive\Documents\IGL DEEP STRUCTURES\PrevRefMatrices\common_references_.csv")
```

```python
df.head()
```

```python
#names of each professor
cols = df.columns

for i in range(len(cols)):
    #skip the first part of row which just has the professor's name
    if i == 0:
        continue

    #Apply the inverse operation which should make bigger similarities have closer distances
    df[cols[i]] = df[cols[i]].map(lambda x: 1 / (x + 1))
```

```python
df_2 = pd.read_csv(r"C:\Users\praje\OneDrive\Documents\IGL DEEP STRUCTURES\PrevRefMatrices\common_references_.csv")

df_2.index

df_2.head()
```

```python
#Make a matrix with the map [sqrt(prof A)sqrt(prof B)] / (Prof A and B + 1)


cols_2 = df_2.columns
rows = df_2.index

for i in range(len(rows)):
    for j in range(len(cols_2)):
        if j == 0:
            continue

        A = np.sqrt(df_2.iloc[i, i+1])
        B = np.sqrt(df_2.iloc[j -1, j])

        C = df_2.iloc[i, j] + 1

        prof = cols_2[j]
        df_2.at[i, prof] = (A + B) / C



        
```
