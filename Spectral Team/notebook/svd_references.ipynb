```python
import pandas as pd
import json
import os
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("out2.csv", index_col= 0)






df = (df - df.mean())

U, S, Vt = np.linalg.svd(df, full_matrices= False)

df.drop("JungeMarius_papers.json", inplace= True)
df.drop("ZaharescuAlexandru_papers.json", inplace= True)
df.drop("BaloghJózsef_papers.json", inplace= True)
df.drop("SongRenming_papers.json", inplace= True)
df.drop("OikhbergTimur_papers.json", inplace= True)
df.drop("BocaFlorin-Petre_papers.json", inplace= True)
df.drop("FordKevinB_papers.json", inplace= True)

proj = df.dot(Vt.T[:, 0:2])

ax = proj.plot.scatter(x = 0, y = 1)

plt.rcParams.update({'font.size': 8})

for i in range(0, len(df.index)):
    plt.text(proj.at[df.index[i], 0], proj.at[df.index[i], 1], df.index[i].split("_")[0])

#find and test with/w0 outliers
#use l1 l2 norms
```
